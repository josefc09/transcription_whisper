```python
import whisper
import os
from pydub import AudioSegment

def convert_audio_to_wav(audio_path):
    """Convierte un archivo de audio a formato WAV."""
    audio = AudioSegment.from_file(audio_path)
    wav_path = audio_path.rsplit(".", 1)[0] + ".wav"
    audio.export(wav_path, format="wav")
    return wav_path

def transcribe_audio(audio_path):
    """Transcribe un archivo de audio a texto usando Whisper."""
    model = whisper.load_model("small")
    result = model.transcribe(audio_path)
    return result["text"]

def save_transcription(text, output_path):
    """Guarda la transcripción en un archivo de texto."""
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(text)

def main(audio_path):
    """Función principal para convertir voz a texto."""
    if not audio_path.endswith(".wav"):
        audio_path = convert_audio_to_wav(audio_path)

    text = transcribe_audio(audio_path)
    output_path = audio_path.rsplit(".", 1)[0] + "_transcription.txt"
    save_transcription(text, output_path)
    print(f"Transcripción guardada en: {output_path}")

if __name__ == "__main__":
    audio_file = "/content/DamianoDavid.mp3"  # Cambia esto por tu archivo de audio
    main(audio_file)
```
